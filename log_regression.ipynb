```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x - logarithm of chances
x = np.arange(-6, 6, 0.1)

#y - probability of a positive outcome
y = np.exp(x) / (1 + np.exp(x))

plt.plot(x, y)
plt.grid()
plt.xlim(-6, 6)
plt.ylim(-0.1, 1.1)
plt.title('Dependence of the probability of a positive outcome on the natural logarithm of chances', fontsize=12)
plt.xlabel('log(odds)', fontsize=12)
plt.ylabel('p', fontsize=12)

plt.show()
plt.close()
```
